```python
def heapify(array, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2

    if l < n and array[i] < array[l]:
        largest = l
    if r < n and array[largest] < array[r]:
        largest = r

    if largest != i:
        array[i], array[largest] = array[largest], array[i]
        heapify(array, n, largest)


def heapSort(array):
    n = len(array)
    for i in range(n // 2 - 1, -1, -1):
        heapify(array, n, i)
        
    for i in range(n - 1, 0, -1):
        array[i], array[0] = array[0], array[i]
        heapify(array, i, 0)

    return array

nums = [9, 7, 6, 15, 16, 5, 10, 11]
heapSort(nums)
print(f"\nОтсортированный массив: {nums}")
```

# Графы

Зада́ча о кёнигсбе́ргских мостах или задача Эйлера — старинная математическая задача, в которой спрашивалось, как можно пройти по всем семи мостам центра старого Кёнигсберга, не проходя ни по одному из них дважды. Впервые была решена в статье, датированной 1736 годом, математиком Леонардом Эйлером, который доказал, что это невозможно, и по ходу доказательства изобрёл эйлеровы циклы. Решение Эйлером задачи о кёнигсбергских мостах явилось первым в истории применением теории графов, но без использования термина «граф» и без рисования диаграмм графов.

Эйлеров граф отличен тем, что в нем можно обойти все вершины и при этом пройти одно ребро только один раз. В нём каждая вершина должна иметь только чётное число рёбер. 

Эйлеров путь (эйлерова цепь) в графе — это путь, проходящий по всем рёбрам графа и притом только по одному разу. 

Эйлеров цикл — эйлеров путь, являющийся циклом, то есть замкнутый путь, проходящий через каждое ребро графа ровно по одному разу.

Полуэйлеров граф — граф, в котором существует эйлеров путь.

Эйлеров граф — граф, в котором существует эйлеров цикл.

### Термины прошлого урока

Смежность вершин - две вершины называются смежными, если они инцидентны одному ребру.

Смежность рёбер - два ребра называются смежными, если они инцедентны одной вершине.

Говоря проще - две вершины смежные, если они соединены ребром, два ребра смежные - если они соединены вершиной.

Петля - ребро, инцидентное одной вершине. Ребро, которое замыкается на одной вершине. 

Псевдограф - граф с петлями. С такими графами не очень удобно работать, потому что переходя по петле мы остаёмся в той же самой вершине, поэтому у него есть своё название

Кратные рёбра - рёбра, имеющие одинаковые концевые вершины, по другому их называют ещё параллельными.

Мультиграф - граф с кратными рёбрами.

Псевдомультиграф - граф с петлями и кратными рёбрами

Степень вершины - это количество рёбер, инцидентных указанной вершине. По-другому - количество рёбер, исходящих из вершины. Петля увеливает степень вершины на 2.

Изолированная вершина - вершина с нулевой степенью.

Висячая вершина - вершина со степенью 1.

Подграф. Если в исходном графе выделить несколько вершин и несколько рёбер (между выбранными вершинами), то мы получим подграф исходного графа. 

Регулярный граф - граф, в котором степени всех вершин одинаковые.

Планарный граф. Если граф можно разместить на плоскости таким образом, чтобы рёбра не пересекались, то он называется “планарным графом” или “плоским графом”.

Путь или Маршрут - это последовательность смежных рёбер. Обычно путь задаётся перечислением вершин, по которым он пролегает.

Длина пути - количество рёбер в пути.

Цепь - маршрут без повторяющихся рёбер.

Простая цепь - цепь без повторяющихся вершин.

Дерево - связный граф без циклов. Между любыми двумя вершинами дерева существует единственный путь.

Цикл или Контур - цепь, в котором последняя вершина совпадает с первой. 

Длина цикла - количество рёбер в цикле.

### Некоторые новые термины

Ориентированный граф или Орграф - граф, в котором рёбра имеют направления.

Дуга - направленные рёбра в ориентированном графе.


Полустепень захода вершины - количество дуг, заходящих в эту вершину.

Исток - вершина с нулевой полустепенью захода.

Полустепень исхода вершины - количество дуг, исходящих из этой вершины

Сток - вершина с нулевой полустепенью исхода.

Компонента связности - множество таких вершин графа, что между любыми двумя вершинами существует маршрут.

Компонента сильной связности - максимальное множество вершин орграфа, между любыми двумя вершинами которого существует путь по дугам.

Компонента слабой связности - максимальное множество вершин орграфа, между любыми двумя вершинами которого существует путь по дугам без учёта направления (по дугам можно двигаться в любом направлении).

```python
def bfs(graph, start, destination):
    visited = set()
    queue = deque([start])
    
    while queue:
        node = queue.popleft()
        if node == destination:
            return True
        if node not in visited:
            visited.add(node)
            queue.extend(graph[node])
    
    return False

def dfs(graph, start, destination, visited=None):
    if visited is None:
        visited = set()
    
    if start == destination:
        return True
    
    if start not in visited:
        visited.add(start)
        for neighbor in graph[start]:
            if dfs(graph, neighbor, destination, visited):
                return True

    return False
```

```python
from collections import defaultdict, deque

def bfs_paths(graph, start, destination):
    queue = deque([(start, [start])])  # Очередь (текущая вершина, путь до неё)
    paths = []  # Список всех найденных путей

    while queue:  # Пока есть элементы в очереди
        node, path = queue.popleft()  # Берём вершину и путь до неё
        if node == destination:  # Если дошли до конца, сохраняем путь
            paths.append(path)

        for neighbor in graph[node]:  # Перебираем всех соседей
            if neighbor not in path:  # Избегаем циклов!
                queue.append((neighbor, path + [neighbor]))  # Добавляем новый путь в очередь вот тут у нас из [0,1] получается [0,1,2]
    
    return paths 

def validPath(n, edges, source, destination):
    graph = defaultdict(list)
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)

    bfs_result = bfs_paths(graph, source, destination)
    for path in bfs_result:
        print(" -> ".join(map(str, path)))

    return len(bfs_result) > 0  # Возвращаем True, если хотя бы один путь найден

print("\nПример 1:")
validPath(3, [[0,1],[1,2],[2,0]], 0, 2)

print("\nПример 2:")
validPath(6, [[0,1],[0,2],[3,5],[5,4],[4,3]], 0, 5)
```

## Поиск кратчайшего пути

Ребра с отрицательным весом — это такие рёбра, при прохождении через которые длина пути как бы уменьшается. Они характерны для экономических задач, в которых выполнение отдельных операций может приносить не только прибыль, но и убытки.


Алгоритм Дейкстры работает для графов, у которых нет рёбер с отрицательным весом. В случае наличия таких рёбер алгоритм Дейкстры нельзя применять, поскольку в ряде случаев он не находит даже приближённых решений. 


Алгоритм Беллмана-Форда, в отличие от алгоритма Дейкстры, допускает рёбра с отрицательным весом. Он находит в ориентированном графе кратчайшие пути от исходной вершины до всех остальных.

### Алгоритм Дейкстры

Трубет список смежности

```python
import heapq
h = [8, 56, 45, 78, 32, 12, 4]
heapq.heapify(h)
print(h[0])
print(h)
heapq.heappop(h)
print(h[0])
print(h)
```

```python
import heapq

def dijkstra(graph, start):
    n = len(graph)
    INF = float('inf')
    dist = {node: INF for node in graph}
    dist[start] = 0
    pq = [(0, start)]  # (расстояние, вершина)

    while pq:
        curr_dist, node = heapq.heappop(pq)

        if curr_dist > dist[node]:
            continue

        for neighbor, weight in graph[node].items():
            distance = curr_dist + weight
            if distance < dist[neighbor]:
                dist[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))
    return dist

# Пример использования
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'C': 2, 'D': 5},
    'C': {'A': 4, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}
start_node = 'A'
print(dijkstra(graph, start_node))
```

### Алгоритм Форда-Беллмана

Требует список ребер

Работает даже с отрицательными рёбрами, но не может работать с отрицательными циклами (обнаруживает их).

Идея:
Начинаем с начальной вершины, присваивая ей расстояние 0, а всем остальным – бесконечность.
Повторяем n-1 раз (где n – число вершин), обновляя расстояния:
Если dist[u] + вес(u → v) < dist[v], то обновляем dist[v].
Дополнительно запускаем 1 проверку: если можно улучшить расстояние, значит есть отрицательный цикл.

```python
def bellman_ford(graph, start):
    INF = float('inf')
    dist = {node: INF for node in graph}
    dist[start] = 0
    edges = [(u, v, w) for u in graph for v, w in graph[u].items()]
    # Обновление расстояний n-1 раз
    for i in range(len(graph) - 1):
        for u, v, w in edges:
            if dist[u] != INF and dist[u] + w < dist[v]:
                dist[v] = dist[u] + w

    # Проверка на отрицательные циклы
    for u, v, w in edges:
        if dist[u] != INF and dist[u] + w < dist[v]:
            raise ValueError("Граф содержит отрицательный цикл")

    return dist

# Пример графа
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'C': -2, 'D': 2},
    'C': {'D': 3},
    'D': {}
}
start_node = 'A'
print(bellman_ford(graph, start_node))
```

### Алгоритм Флойда-Уоршелла

Работает для нахождения кратчайших путей между всеми парами вершин. Может обрабатывать отрицательные веса рёбер (но не работает с отрицательными циклами).

Идея:
Создаём матрицу расстояний, где dist[i][j] – вес ребра i → j, если оно существует, иначе INF.
Последовательно пробуем улучшить путь i → j через промежуточную вершину k.
Если dist[i][j] > dist[i][k] + dist[k][j], то обновляем dist[i][j].

```python
def floyd_warshall(graph):
    INF = float('inf')
    nodes = list(graph.keys())
    n = len(nodes)

    # Инициализация матрицы расстояний
    dist = {i: {j: INF for j in nodes} for i in nodes}
    for i in nodes:
        dist[i][i] = 0
        for j, w in graph[i].items():
            dist[i][j] = w

    # Основной алгоритм
    for k in nodes:
        for i in nodes:
            for j in nodes:
                if dist[i][j] > dist[i][k] + dist[k][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]

    return dist

graph = {
    'A': {'B': 3, 'C': 8},
    'B': {'A': 3, 'C': 2, 'D': 5},
    'C': {'A': 8, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}
print(floyd_warshall(graph))
```

```python
from collections import defaultdict, deque

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = defaultdict(list)

    def add_edge(self, u, v):
        self.graph[u].append(v)

    def _dfs(self, v, visited, stack):
        visited[v] = True
        for neighbor in self.graph[v]:
            if not visited[neighbor]:
                self._dfs(neighbor, visited, stack)
        stack.append(v)

    def _transpose(self):
        transposed_graph = Graph(self.V)
        for node in self.graph:
            for neighbor in self.graph[node]:
                transposed_graph.add_edge(neighbor, node)
        return transposed_graph

    def _fill_order(self, visited, stack):
        for i in range(self.V):
            if not visited[i]:
                self._dfs(i, visited, stack)

    def _dfs_util(self, v, visited, component):
        visited[v] = True
        component.append(v)
        for neighbor in self.graph[v]:
            if not visited[neighbor]:
                self._dfs_util(neighbor, visited, component)

    def kosaraju_scc(self):
        stack = deque()
        visited = [False] * self.V

        self._fill_order(visited, stack)

        transposed_graph = self._transpose()

        visited = [False] * self.V
        scc_list = []

        while stack:
            node = stack.pop()
            if not visited[node]:
                component = []
                transposed_graph._dfs_util(node, visited, component)
                scc_list.append(component)

        return scc_list

if __name__ == "__main__":
    g = Graph(5)
    g.add_edge(1, 0)
    g.add_edge(0, 2)
    g.add_edge(2, 1)
    g.add_edge(0, 3)
    g.add_edge(3, 4)

    sccs = g.kosaraju_scc()
    print("Strongly Connected Components:", sccs)
```

## Топологическая сортировка

Топологическая сортировка для ориентированного ациклического графа (Directed Acyclic Graphs, далее DAG) — это линейное упорядочение вершин, для которого выполняется следующее условие — для каждого направленного ребра uv вершина u предшествует вершине v в упорядочении. Если граф не является DAG, то топологическая сортировка для него невозможна.

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D'],
    'C': ['D'],
    'D': []
}
```

```python
def topological_sort_dfs(graph):
    visited = set()  #Создаём множество, чтобы не заходить в одну вершину дважды
    stack = [] # будем добавлять вершины после посещения всех их соседей

    def dfs(node):
        if node in visited:
            return  #Если уже в visited, значит, обработали её ранее, поэтому возвращаемся
        visited.add(node)
        
        for neighbor in graph[node]:
            dfs(neighbor)
        
        stack.append(node)  # Добавляем в стек после посещения всех соседей

    for node in graph:
        if node not in visited:
            dfs(node)

    return stack[::-1]  # Разворачиваем стек (порядок должен быть обратным)
print(topological_sort_dfs(graph))
```

Каждыи‌ маршрут (скажем, i-и‌) соединяет две станции ui и vi, электрички по нему ездят с некоторои‌ частотои‌ fi | f (то есть fi является делителем некоторого числа f), первая из них отправляется в момент времени si < fi, каждая едет ti минут, а проезд стоит ci рублеи‌. При этом запрещено стоять на месте (то есть ждать электричку нельзя, можно только быстро пересесть). Определите, за какое минимальное время можно попасть из 1- го города в n-и‌ и сообщите минимальное количество потраченных рублеи‌ в таком случае. Путешествие начинается в момент времени 0. Асимптотика: O(f m · log(f n)).

## Задача

Задача 2, как к существующему решению добавить теорию графов?)

Задан список из n городов. Название каждого города представляет собой строку, состоящую из маленьких латинских букв. За линейное время от размера входных данных определите, существует ли такая перестановка городов, что первая буква i-го города равна последней букве (i − 1)-го для всех подходящих i.

```python
from collections import Counter

def can_rearrange_cities(cities):
    first_letters = Counter()  # Количество вхождений букв в начале городов
    last_letters = Counter()   # Количество вхождений букв в конце городов
    
    # Подсчёт частоты первой и последней букв
    for city in cities:
        first_letters[city[0]] += 1
        last_letters[city[-1]] += 1

    # Подсчитываем "дисбаланс" букв (сколько раз буква встречается больше в начале или в конце)
    start_excess = 0
    end_excess = 0
    
    for letter in set(first_letters.keys()).union(last_letters.keys()):
        diff = first_letters[letter] - last_letters[letter]
        if diff == 1:
            start_excess += 1
        elif diff == -1:
            end_excess += 1
        elif diff != 0:  # Если разница больше 1 в любую сторону – цепочку построить нельзя
            return False

    # Допустимы два случая:
    # 1. Все буквы сбалансированы (start_excess == 0 и end_excess == 0)
    # 2. Есть ровно одна "начальная" лишняя буква и одна "конечная" (start_excess == 1 и end_excess == 1)
    return (start_excess == 0 and end_excess == 0) or (start_excess == 1 and end_excess == 1)


cities1 = ["london", "newyork", "klin", "vienna"]  
cities2 = ["paris", "sydney", "york", "kyoto"]  
cities3 = ["paris", "london", "moscow"] 
cities4 = ["a", "b", "c"]  

print(can_rearrange_cities(cities1)) 
print(can_rearrange_cities(cities2)) 
print(can_rearrange_cities(cities3))
```
