### Аппроксимация

```python
from scipy.optimize import curve_fit

x = np.array([1, 2, 3, 4, 5])
y = np.array([20, 30, 60, 50, 60])

def func(x, a, b):
    return a * x + b

popt, pcov = curve_fit(func, x, y)
#popt: массив, содержащий оптимальные значения параметров a и b линейной функции, которые лучше всего аппроксимируют данные.
#pcov: матрица ковариации, которая содержит информацию о погрешностях параметров a и b.

print("Коэффициенты аппроксимации:")
print("a =", popt[0])
print("b =", popt[1])

plt.scatter(x, y, label='Данные')
plt.plot(x, func(x, *popt), label='Аппроксимация', color='red')
plt.legend()
plt.show()
```

```python
def sum_it(n):
    x = 0
    for n in range(1, n+1):
        x += n
    return x

sum_it(10)
```

```python
def sum_rec(n):
    if n == 1:
        return 1
    return n + sum_rec(n-1)

sum_rec(10)
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
def Max(A):
    if len(A) == 1:
        return A[0]
    else:
        return max(Max(A[:-1]), A[-1])
```

```python
def power(a, n):
    if n == 0:
        return 1
    elif n % 2 == 1:
        return power(a, n - 1) * a
    else:
        return power(a, n // 2) ** 2
```

Ханойский башни

```python
n = int(input())
def play_hanoi(n, from_, to_, using_what_):
    """
    from_ стержень с которого перекладываем диски
    to_ стержень на который перекладываем диски
    using_what_ стержень на который перекладываем для временного хранения
    """
    if n <= 0: return
    play_hanoi(n-1, from_, using_what_, to_)
    print("Диск ", n, " : ", from_, " --> ", to_)
    play_hanoi(n-1, using_what_, to_, from_)

play_hanoi (n, '1', '2', '3')
```

Треугольник Паскаля

Напишите функцию pascal(n), которая будет принимать положительные целые числа n и возвращать n-ую строку треугольника Паскаля.

```python
def pascal(n):
    if n == 1:
        return [1]
    else:
        prev_row = pascal(n-1)
        row = [x + y for x, y in zip([0] + prev_row, prev_row + [0])]
        return row

print(pascal(5)) 
```

Размен монет

```python
def  make_exchange(left_money, coins):
    if len(coins) < 1 or left_money < 0:
        return 0
    if left_money == 0:
        return 1

    coins = list(reversed(sorted(coins)))
    new_coins = coins.copy()
    new_coins.pop(0)
    s = make_exchange(left_money - coins[0], coins) + make_exchange(left_money, new_coins)

    return(s)
```

```python
from random import randint

WIDTH = 300
HEIGHT = 200


class Ball:
    def __init__(self):
        self.R = randint(10, 50) #храним размер, при каждом создании объекта будет выбираться случайно
        self.x = randint(self.R, WIDTH - self.R) # храним положение по x и y
        self.y = randint(self.R, HEIGHT - self.R)
        self.dx, self.dy = (10, 10) # это по сути шаг движения шаров. если увеличить -- будут двигаться быстрее
        self.ball_id = canvas.create_oval(self.x - self.R,
                                     self.y - self.R,
                                     self.x + self.R,
                                     self.y + self.R, fill="green") # при создании шарика отрисовываем его

    def move(self):
        self.x += self.dx
        self.y += self.dy
        if self.x + self.R > WIDTH or self.x - self.R <= 0: # отражение от стенок
            self.dx = -self.dx
        if self.y + self.R > HEIGHT or self.y - self.R <= 0: # отр
            self.dy = -self.dy

    def show(self):
        canvas.move(self.ball_id, self.dx, self.dy)


def click_handler(event):
    print('Hello World! x=', event.x, 'y=', event.y)

#здесь мы уже привычно обращаемся к balls как к глобальной переменной. На самом деле дело в том, что нам лень писать классы.
def tick():
    for ball in balls:
        ball.move()
        ball.show()
    root.after(50, tick)


root = Tk()
root.geometry(f'{WIDTH}x{HEIGHT}')
canvas = Canvas(root)
canvas.pack()
#сделаем так, чтобы нажатие левой кнопки на поле выводило координаты точки, в которую мы нажали
canvas.bind('<Button-1>', click_handler)
balls = [Ball() for i in range(5)]
# делаем шаг перемещения и отрисовки шаров. поскольку mainloop циклит наше приложение, это будет происходить, пока мы не закроем окно
tick()
root.mainloop()
```
